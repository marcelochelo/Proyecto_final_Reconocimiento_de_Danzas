```
#librerias
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```
#carpeta de caporales
caporales_folder_path ="/content/caporales"
caporales=[]
img_size=150

for img in os.listdir(caporales_folder_path):
  img = cv2.imread(os.path.join(caporales_folder_path,img))
  #img_gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img_resize = cv2.resize(img,(img_size,img_size))
  caporales.append(img_resize)
```

```
caporales = np.array(caporales)
print(caporales.shape)
```

```
#carpeta de tinkuy
tinkuy_folder_path ="/content/tinkuy"
tinkuy=[]
img_size=150

for img in os.listdir(tinkuy_folder_path):
  img = cv2.imread(os.path.join(tinkuy_folder_path,img))
  #img_gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img_resize = cv2.resize(img,(img_size,img_size))
  tinkuy.append(img_resize)
```

```
tinkuy = np.array(tinkuy)
print(tinkuy.shape)
```

```
#visualizar alguna imagen
#este caso de tinkuy
print(tinkuy[4].shape)
plt.figure()
plt.imshow(np.squeeze(tinkuy[5]))
plt.colorbar()
plt.grid(False)
plt.show()
```

```
#concatenar 
#total de las imagens
images = np.concatenate([caporales,tinkuy])
print(len(images))
#converit en arreglo
Images = np.array(images)
print(Images.shape)
```

```
#relacionar un numero con imagen caporales con el numero 0
etiquetas_caporales = np.repeat(0,197)
print(len(etiquetas_caporales))
print (etiquetas_caporales)
```

```
#relacionar un numero con imagen tinkuy con el numero 1
etiquetas_tinkuy = np.repeat(1,199)
print(len(etiquetas_tinkuy))
print (etiquetas_tinkuy)
```

```
#asignamos la clase con el numero
# 0 caporales
#1 tinkuy
class_name=(['Caporales','Tinkuy'])
```

```
#juntar todas las etiquetas, unimos todos
labels = np.concatenate([etiquetas_caporales,etiquetas_tinkuy])
print(len(labels))
print(labels)
Labels=np.array(labels)
print(Labels.shape)
```

```
#visualizar de manera ordenada las etiquetas
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(Images[i])
    #, cmap=plt.cm.binary
    plt.xlabel(class_name[Labels[i]])
plt.show()
```

```
from __future__ import absolute_import, division, print_function, unicode_literals

variable_name = ""
try:
  # %tensorflow_version only exists in Colab.
  %tensorflow_version 2.x
except Exception:
  pass

from __future__ import absolute_import, division, print_function, unicode_literals

# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers, models
import tensorflow.keras.optimizers as Optimizer

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
```

```
#entrenamiento
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(150, 150,3)),
    keras.layers.Dense(150, activation='relu'),
    
    keras.layers.Dense(2, activation='softmax'),
    
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(Images, Labels, epochs=30)
trained=model.fit(Images, Labels, epochs=30)
```

```
#seleccionar una imagen del arreglo completo
img=Images[230]
print(img.shape)
img = (np.expand_dims(img,0))
print(img.shape)
```

```
#hacer la prediccion de la imagen
prediction_single = model.predict(img)
print(prediction_single)
print(np.sum(prediction_single))
print(np.argmax(prediction_single))
print(class_name[np.argmax(prediction_single)])
```
