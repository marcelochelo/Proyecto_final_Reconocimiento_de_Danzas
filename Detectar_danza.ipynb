```
#librerias
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```
#carpeta de caporales
caporales_folder_path ="/content/caporales"
caporales=[]
img_size=150

#devuelve la carpeta especificada contiene un nombre de carpeta
for img in os.listdir(caporales_folder_path):
  img = cv2.imread(os.path.join(caporales_folder_path,img))
  #img_gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img_resize = cv2.resize(img,(img_size,img_size))
  caporales.append(img_resize)
```

```
caporales = np.array(caporales)
print(caporales.shape)
```

```
#carpeta de diablada
diablada_folder_path ="/content/diablada"
diablada=[]
img_size=150

for img in os.listdir(diablada_folder_path):
  img = cv2.imread(os.path.join(diablada_folder_path,img))
  #img_gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img_resize = cv2.resize(img,(img_size,img_size))
  diablada.append(img_resize)
```

```
diablada = np.array(diablada)
print(diablada.shape)
```

```
#carpeta de morenada
morenada_folder_path ="/content/morenada"
morenada=[]
img_size=150

for img in os.listdir(morenada_folder_path):
  img = cv2.imread(os.path.join(morenada_folder_path,img))
  #img_gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img_resize = cv2.resize(img,(img_size,img_size))
  morenada.append(img_resize)
```

```
morenada = np.array(morenada)
print(morenada.shape)
```

```
#carpeta de salay
salay_folder_path ="/content/salay"
salay=[]
img_size=150

for img in os.listdir(salay_folder_path):
  img = cv2.imread(os.path.join(salay_folder_path,img))
  #img_gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img_resize = cv2.resize(img,(img_size,img_size))
  salay.append(img_resize)
```

```
salay = np.array(salay)
print(salay.shape)
```

```
#carpeta de tinkuy
tinkuy_folder_path ="/content/tinkuy"
tinkuy=[]
img_size=150

for img in os.listdir(tinkuy_folder_path):
  img = cv2.imread(os.path.join(tinkuy_folder_path,img))
  #img_gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img_resize = cv2.resize(img,(img_size,img_size))
  tinkuy.append(img_resize)
```

```
tinkuy = np.array(tinkuy)
print(tinkuy.shape)
```

```
#visualizar alguna imagen
#este caso de salay
print(salay[6].shape)
plt.figure()
plt.imshow(np.squeeze(salay[6]))
plt.colorbar()
plt.grid(False)
plt.show()
```

```
#concatenar 
#total de las imagens
images = np.concatenate([caporales,diablada,morenada,salay,tinkuy])
print(len(images))

#converit en arreglo
Images = np.array(images)
print(Images.shape)
```

```
#relacionar un numero con imagen caporales con el numero 0.2
etiquetas_caporales = np.repeat(0.2,197)
print(len(etiquetas_caporales))
print (etiquetas_caporales)
```

```
#relacionar un numero con imagen diablada con el numero 0.4
etiquetas_diablada = np.repeat(0.4,196)
print(len(etiquetas_diablada))
print (etiquetas_diablada)
```

```
#relacionar un numero con imagen morenada con el numero 0.6
etiquetas_morenada = np.repeat(0.6,199)
print(len(etiquetas_morenada))
print (etiquetas_morenada)
```

```
#relacionar un numero con imagen salay con el numero 0.8
etiquetas_salay = np.repeat(0.8,198)
print(len(etiquetas_salay))
print (etiquetas_salay)
```

```
#relacionar un numero con imagen tinkuy con el numero 1
etiquetas_tinkuy = np.repeat(1,200)
print(len(etiquetas_tinkuy))
print (etiquetas_tinkuy)
```

```
#nombre de las posiciones 
#0.2 =caporales
#0.4 =diablada
#0.6 =morenada
#0.8 =salay
#1 =tinkuy
class_name=(['Caporales','Diablada','Morenada','Salay','Tinkuy'])
```

```
#juntar todas las etiquetas
labels = np.concatenate([etiquetas_caporales,etiquetas_diablada,etiquetas_morenada,etiquetas_salay,etiquetas_tinkuy])
print(len(labels))
print(labels)
Labels=np.array(labels)
print(Labels.shape)
```
